## Types of Machine Learning

Machine Learning is broadly divided into four main types:

### 1. **Supervised Learning**
- The model learns from **labeled data** ‚Äî examples where both input (X) and output (y) are known.  
- Goal: learn a mapping function \( f: X \to y \)
- Examples:  
  - **Regression:** Predicting house prices, sales, temperature  
  - **Classification:** Spam detection, disease prediction  

üîπ *Linear Regression is the simplest supervised learning algorithm.*

---

### 2. **Unsupervised Learning**
- The data is **unlabeled** ‚Äî only inputs (X) are given, no target output.  
- Goal: find hidden patterns or structures.  
- Examples: PCA, K-Means Clustering, Dimensionality Reduction  

---

### 3. **Semi-Supervised Learning**
- Combination of labeled and unlabeled data.  
- Common in real-world scenarios where labeling is expensive (e.g., medical images).  

---

### 4. **Reinforcement Learning**
- The model learns by **trial and error**, receiving rewards or penalties.  
- Examples: game-playing AI, robotics, self-driving cars.

---

## Learning Paradigms ‚Äî How Models Learn

### üî∏ Batch (Offline) Learning
- Model is trained on the **entire dataset at once**.  
- After training, it doesn‚Äôt learn anymore until retrained with new data.  
- ‚úÖ Good for static data  
- ‚ùå Bad for continuously changing environments  

### üî∏ Online Learning
- Model learns **incrementally** as new data comes in.  
- ‚úÖ Good for streaming or real-time data (like stock prediction)  
- ‚ùå Risk of forgetting older patterns  

### üî∏ Out-of-Core Learning
- Handles datasets **too large to fit in memory** by loading them in chunks.  
- Common in big data pipelines.  

---

## Batch vs Online Learning ‚Äî Quick Comparison

| Feature | Batch Learning | Online Learning |
|----------|----------------|----------------|
| Data size | Fits in memory | Streams over time |
| Training time | High | Incremental |
| Adaptability | Static | Dynamic |
| Example | Linear Regression on CSV | Stock price prediction in real time |

---

## Instance-Based vs Model-Based Learning

###  Instance-Based Learning
- Model memorizes the data and compares new examples to stored ones.  
- Example: K-Nearest Neighbors (KNN)  
- Analogy: ‚ÄúRemember and compare.‚Äù

### Model-Based Learning
- Model **learns a function** that maps input to output.  
- Example: Linear Regression  
- Analogy: ‚ÄúUnderstand the trend and generalize.‚Äù

| Type | Approach | Example |
|------|-----------|----------|
| Instance-Based | Memorize and compare | KNN |
| Model-Based | Learn a function | Linear Regression |

# Simple Linear Regression (SLR)

Now that we understand where Linear Regression fits into the world of Machine Learning,  
let‚Äôs dive into **Simple Linear Regression**, or SLR for short.

---

## üí°Intuition

The goal of Linear Regression is simple:
> Find the best straight line that explains how one variable depends on another.

In other words:
- We have one **input feature** (X)
- And one **output/target** (y)
- We want to find the line that best fits all the data points.

---

### Example

Imagine we have data about the **number of study hours (X)** and the **exam score (y)** of students.

| Hours (X) | Score (y) |
|------------|-----------|
| 1 | 45 |
| 2 | 50 |
| 3 | 60 |
| 4 | 65 |
| 5 | 80 |

If we plot these points, they‚Äôll roughly follow a straight trend.  
We want a line that can **predict the score** for any given number of hours.

---

## The Linear Regression Equation

The equation of a straight line is:

$$
y = m x + b
$$

where:  
- \( y \) ‚Üí predicted value  
- \( x \) ‚Üí input feature  
- \( m \) ‚Üí slope of the line (how steep it is)  
- \( b \) ‚Üí intercept (where the line crosses the y-axis)  

Our job is to **find the best values of \( m \) and \( b \)** that minimize prediction error.

---

## üéØ Goal ‚Äî Minimizing Error
For each data point:

$$
(x_i, y_i)
$$

the **predicted value** is:

$$
\hat{y}_i = m x_i + b
$$

The **error (residual)** is the difference between actual and predicted values:

$$
e_i = y_i - \hat{y}_i = y_i - (m x_i + b)
$$

We want the line that **minimizes the total squared error**:

$$
\text{Error} = \sum_{i=1}^{n} (y_i - \hat{y}_i)^2
= \sum_{i=1}^{n} (y_i - m x_i - b)^2
$$

> This is called the **Least Squares Criterion**,  
> and the method we use to solve it is called **Ordinary Least Squares (OLS)**.

---

## Why ‚ÄúLeast Squares‚Äù?

Because we **square the errors** (to make them positive and penalize large errors more heavily),  
and we find the values of \( m \) and \( b \) that make this sum as small as possible.

So the optimization problem becomes:

$$
\min_{m,b} \; J(m,b) = \sum_{i=1}^{n} (y_i - m x_i - b)^2
$$

---

## üîç Finding \( m \) and \( b \) ‚Äî Mathematical Derivation

To minimize the error, we take **partial derivatives** of the cost function with respect to \( m \) and \( b \),  
set them equal to zero, and solve for \( m \) and \( b \).

---

### Step 1: Start with the Cost Function

$$
J(m,b) = \sum_{i=1}^{n} (y_i - m x_i - b)^2
$$

---

### Step 2: Differentiate with Respect to \( m \)

$$
\frac{\partial J}{\partial m} = -2 \sum_{i=1}^{n} x_i (y_i - m x_i - b)
$$

---

### Step 3: Differentiate with Respect to \( b \)

$$
\frac{\partial J}{\partial b} = -2 \sum_{i=1}^{n} (y_i - m x_i - b)
$$

---
### Step 4: Set Both Derivatives Equal to Zero

This gives two equations:

$$
\sum x_i y_i = m \sum x_i^2 + b \sum x_i
$$

$$
\sum y_i = m \sum x_i + n b
$$


---
### Step 5: Solve for \( m \) and \( b \)

After simplification, we get:

$$
m = \frac{\sum (x_i - \bar{x})(y_i - \bar{y})}
{\sum (x_i - \bar{x})^2}
$$

$$
b = \bar{y} - m \bar{x}
$$

where:  

$$
\bar{x} = \frac{1}{n}\sum x_i, \quad
\bar{y} = \frac{1}{n}\sum y_i
$$

‚úÖ These are the **closed-form formulas** for the best-fitting line.

---


## Intuitive Interpretation

- The **slope (m)** tells us how much \( y \) changes when \( x \) increases by 1.  
  - If \( m = 2 \), then every 1-hour increase in study time increases the score by 2 points.
- The **intercept (b)** tells us what \( y \) would be when \( x = 0 \).  
  - For example, the expected score without studying.

---

## Geometric Intuition

Imagine rotating a straight line through the cloud of data points until the **sum of vertical distances (errors)** is as small as possible.  
That‚Äôs what OLS does ‚Äî it finds the line that‚Äôs ‚Äúclosest‚Äù to all points overall.

---

‚úÖ **Up Next ‚Üí Code Implementation**

Now that we know the math behind Linear Regression,  
let‚Äôs move to **Step 4 ‚Äî Implementing Simple Linear Regression from Scratch using NumPy**.


```python
# Simple Linear Regression (NumPy Implementation)
import numpy as np
import matplotlib.pyplot as plt

# Step 1: Create sample data
# Hours studied (X) and corresponding exam scores (y)
X = np.array([1, 2, 3, 4, 5])
y = np.array([45, 50, 60, 65, 80])

# Number of samples
n = len(X)

# Step 2: Compute means
x_mean = np.mean(X)
y_mean = np.mean(y)

# Step 3: Compute slope (m) and intercept (b) using OLS formulas
# m = Œ£(x - xÃÑ)(y - »≥) / Œ£(x - xÃÑ)¬≤
numerator = np.sum((X - x_mean) * (y - y_mean))
denominator = np.sum((X - x_mean)**2)
m = numerator / denominator

# b = »≥ - m xÃÑ
b = y_mean - (m * x_mean)

print(f"Slope (m): {m:.3f}")
print(f"Intercept (b): {b:.3f}")

# Step 4: Make predictions
y_pred = m * X + b

# Step 5: Visualization
plt.figure(figsize=(8,6))
plt.scatter(X, y, color='blue', label='Actual Data Points')
plt.plot(X, y_pred, color='red', label='Best Fit Line')
plt.title('Simple Linear Regression (From Scratch)')
plt.xlabel('Study Hours')
plt.ylabel('Exam Score')
plt.legend()
plt.grid(True)
plt.show()
```

# Understanding the Implementation

### Step-by-Step Summary

---

### 1Ô∏è Data Preparation  
We start with 5 data points (Hours studied ‚Üí Exam score).  
The data roughly follows a linear pattern, suggesting a straight-line relationship.

---

### 2Ô∏è Calculate Means  
We find the average (mean) of both X and y:

$$
\bar{x} = \frac{1}{n}\sum_{i=1}^{n} x_i, \quad
\bar{y} = \frac{1}{n}\sum_{i=1}^{n} y_i
$$

---

### 3Ô∏è Compute Slope (m)

Using the **Ordinary Least Squares (OLS)** formula:

$$
m = \frac{\sum_{i=1}^{n} (x_i - \bar{x})(y_i - \bar{y})}
{\sum_{i=1}^{n} (x_i - \bar{x})^2}
$$

The slope \(m\) shows **how much y changes** when x increases by 1.

---

### 4Ô∏è Compute Intercept (b)

The intercept represents the expected value of \(y\) when \(x = 0\):

$$
b = \bar{y} - m\bar{x}
$$

---

### 5Ô∏è Prediction Formula

Once \(m\) and \(b\) are known, we can predict new values of \(y\):

$$
\hat{y} = m x + b
$$

Here,  
- \( \hat{y} \) = predicted value  
- \( x \) = input feature  
- \( m \) = slope  
- \( b \) = intercept  

---

### 6Ô∏è Visualization  

The red line in our plot represents the **best-fit line** that minimizes the total squared error between predicted and actual values.

This line passes **as close as possible** to all the data points, balancing the distances above and below it.

---

‚úÖ **Interpretation Recap**

- **Positive Slope (m):** as x increases, y increases.  
- **Intercept (b):** predicted y when x = 0.  
- **Best Fit Line:** minimizes the total squared error (Least Squares method).  

---

**Mathematically**, the total error minimized by our model is:

$$
J(m,b) = \sum_{i=1}^{n} (y_i - (m x_i + b))^2
$$

The smaller \( J(m,b) \), the better the fit.


# Regression Evaluation Metrics  

After fitting our regression line, we need to know **how well our model performs**.  
Evaluation metrics help us measure **how close our predictions are** to the actual values.

Let‚Äôs go through the most important metrics one by one

---

## 1Ô∏è Mean Absolute Error (MAE)

MAE measures the **average magnitude of errors** ‚Äî  
how far, on average, the predictions are from the actual values.

$$
\text{MAE} = \frac{1}{n} \sum_{i=1}^{n} |y_i - \hat{y}_i|
$$

- It‚Äôs easy to interpret ‚Äî "on average, how wrong were we?"  
- Doesn‚Äôt penalize large errors heavily.  

‚úÖ **Lower MAE ‚Üí Better model**

---

## 2Ô∏è Mean Squared Error (MSE)

MSE measures the **average squared difference** between actual and predicted values.

$$
\text{MSE} = \frac{1}{n} \sum_{i=1}^{n} (y_i - \hat{y}_i)^2
$$

- Squaring makes sure all errors are positive.  
- Penalizes **larger errors** more than smaller ones.  

‚úÖ **Lower MSE ‚Üí Better model**

---

## 3Ô∏è Root Mean Squared Error (RMSE)

RMSE is simply the square root of MSE ‚Äî it brings the error back to the same units as the target variable.

$$
\text{RMSE} = \sqrt{\frac{1}{n} \sum_{i=1}^{n} (y_i - \hat{y}_i)^2}
$$

- Easier to interpret because it‚Äôs in the same unit as \(y\).  
- Sensitive to large errors (since it‚Äôs derived from MSE).  

‚úÖ **Lower RMSE ‚Üí Better model**

---

## 4Ô∏è Coefficient of Determination (R¬≤ Score)

R¬≤ measures **how much of the variance** in the dependent variable \(y\)  
is explained by the model.

$$
R^2 = 1 - \frac{\sum (y_i - \hat{y}_i)^2}
{\sum (y_i - \bar{y})^2}
$$

- Numerator = **Residual Sum of Squares (SSR)**  
- Denominator = **Total Sum of Squares (SST)**  
- \( R^2 = 1 \) ‚Üí perfect prediction  
- \( R^2 = 0 \) ‚Üí model predicts no better than mean  

‚úÖ **Higher R¬≤ ‚Üí Better model**

---

## 5Ô∏è Adjusted R¬≤

R¬≤ tends to **increase artificially** as we add more predictors (even if they don‚Äôt help).  
Adjusted R¬≤ fixes this by penalizing unnecessary features.

$$
R^2_{adj} = 1 - (1 - R^2) \times \frac{n - 1}{n - p - 1}
$$

Where:  
- \( n \) = number of samples  
- \( p \) = number of predictors (features)

‚úÖ **Higher Adjusted R¬≤ ‚Üí Better model (while considering model simplicity)**

---

## üß† Intuitive Summary

| Metric | Formula | Best Value | Interpretation |
|:--------|:---------|:-------------|:----------------|
| **MAE** | \( \frac{1}{n}\sum|y-\hat{y}| \) | 0 | Avg absolute error |
| **MSE** | \( \frac{1}{n}\sum(y-\hat{y})^2 \) | 0 | Penalizes large errors |
| **RMSE** | \( \sqrt{\frac{1}{n}\sum(y-\hat{y})^2} \) | 0 | Same units as y |
| **R¬≤** | \( 1 - \frac{SSR}{SST} \) | 1 | Variance explained |
| **Adj. R¬≤** | \( 1 - (1 - R^2)\frac{n-1}{n-p-1} \) | 1 | Adjusted for extra features |

---

‚úÖ **In short:**  
- Use **MAE** when you care about *average* error size.  
- Use **MSE/RMSE** when you want to penalize large mistakes.  
- Use **R¬≤** and **Adjusted R¬≤** to measure *overall fit and explanatory power*.

---

Next, we‚Äôll **implement these metrics in NumPy** and test them on our regression model.

```python
# Implementing Regression Metrics from Scratch (NumPy)

# Actual vs Predicted from previous regression
y_actual = y
y_predicted = y_pred
n = len(y_actual)

# 1Ô∏è Mean Absolute Error
MAE = np.mean(np.abs(y_actual - y_predicted))

# 2Ô∏è Mean Squared Error
MSE = np.mean((y_actual - y_predicted) ** 2)

# 3Ô∏è Root Mean Squared Error
RMSE = np.sqrt(MSE)

# 4Ô∏è R¬≤ Score
SS_Residual = np.sum((y_actual - y_predicted) ** 2)
SS_Total = np.sum((y_actual - np.mean(y_actual)) ** 2)
R2 = 1 - (SS_Residual / SS_Total)

# 5Ô∏è Adjusted R¬≤ Score
p = 1  # number of predictors (for simple linear regression)
Adjusted_R2 = 1 - (1 - R2) * ((n - 1) / (n - p - 1))

# Display results
print(f"Mean Absolute Error (MAE): {MAE:.3f}")
print(f"Mean Squared Error (MSE): {MSE:.3f}")
print(f"Root Mean Squared Error (RMSE): {RMSE:.3f}")
print(f"R¬≤ Score: {R2:.3f}")
print(f"Adjusted R¬≤ Score: {Adjusted_R2:.3f}")
```

```python
# visualize prediction errors
plt.figure(figsize=(8,5))
plt.scatter(X, y, color='blue', label='Actual')
plt.plot(X, y_pred, color='red', label='Predicted')
plt.title('Model Fit')
plt.xlabel('X')
plt.ylabel('y')
plt.legend()
plt.grid(True)
plt.show()
```
